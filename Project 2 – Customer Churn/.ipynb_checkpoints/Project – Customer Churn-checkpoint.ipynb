### C) Linear Regression: 
a. Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’

i. Divide the dataset into train and test sets in 70:30 ratio. 

ii. Build the model on train set and predict the values on test set 

iii. After predicting the values, find the root mean square error 

iv. Find out the error in prediction & store the result in ‘error’

v. Find the root mean square error

```python
from sklearn.model_selection import train_test_split
x=pd.DataFrame(df['tenure'])
y=df['MonthlyCharges']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size= 0.7,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

LR =LinearRegression()
LR.fit(x_train,y_train)
```

```python
predict = LR.predict(x_test)
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
mse= mean_squared_error(predict, y_test)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(predict,y_test)

#Root Mean Square Error 
RMSE = np.sqrt(MSE)
RMSE
```

#### D) Logistic Regression: 
a. Build a simple logistic regression modelwhere dependent variable is ‘Churn’ & 
independent variable is ‘MonthlyCharges’

i. Divide the dataset in 65:35 ratio 
    
ii. Build the model on train set and predict the values on test set 

iii. Build the confusion matrix and get the accuracy score 

```python
from sklearn.model_selection import train_test_split
x =pd.DataFrame(df['MonthlyCharges'])
y = df['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size= 0.65,random_state =10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

LOR =LogisticRegression()
LOR.fit(x_train,y_train)
```

```python
predict = LOR.predict(x_test)
predict
```

```python
confusion_matrix(predict,y_test)

# TP FP
# FN TN
```

```python
accuracy_score(predict,y_test)
```

#### b. Build a multiple logistic regression model where dependent variable is ‘Churn’ & 
independent variables are ‘tenure’ & ‘MonthlyCharges’

i. Divide the dataset in 80:20 ratio 
    
ii. Build the model on train set and predict the values on test set 

iii. Build the confusion matrix and get the accuracy score 


```python
x=pd.DataFrame(df.loc[:,['MonthlyCharges','tenure']])
y=df['Churn']
```

```python
y_predict=LoR.predict(x_test)
y_predict
```

```python
confusion_matrix(y_predict, y_test), accuracy_score(y_predict, y_test)
```

### E) Decision Tree: 
a. Build a decision tree model where dependent variable is ‘Churn’ & independent 
variable is ‘tenure’

i. Divide the dataset in 80:20 ratio 
    
ii. Build the model on train set and predict the values on test set

iii. Build the confusion matrix and calculate the accuracy

```python
from sklearn.model_selection import train_test_split
x=pd.DataFrame(df['tenure'])
y=df['Churn']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state= 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTree= DecisionTreeClassifier()
DecisionTree.fit(x_train, y_train)
```

```python
y_predict=DecisionTree.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_predict, y_test), accuracy_score(y_predict, y_test)
```

#### F) Random Forest: 
a. Build a Random Forest model where dependent variable is ‘Churn’ & independent 
variables are ‘tenure’ and ‘MonthlyCharges’

i. Divide the dataset in 70:30 ratio 
    
ii. Build the model on train set and predict the values on test set 

iii. Build the confusion matrix and calculate the accuracy

```python
x=pd.DataFrame(df.loc[:,['MonthlyCharges','tenure']])
y=df['Churn']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state= 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=100)
RFC.fit(x_train, y_train)
```

```python
y_predict=RFC.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_predict, y_test), accuracy_score(y_predict, y_test)
```
